Before we can proceed with training our model, we need to clean the data by handling missing values, outliers, and other anomalies. This may include removing or imputing missing values, removing outliers, and normalizing the data.
- We convert the 'Date' column to datetime format using pd.to_datetime() and set it as the index of the dataframe using data.set_index(). We also sort the dataframe by date using data.sort_index().

```python
# Check for missing values
df.isnull().sum()

# Remove rows with missing values
df = df.dropna()

# Removing duplicates
df.drop_duplicates(inplace=True)

# Normalize the data
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
df.sort_index(inplace=True)
df = df[['Close']]

# Handling missing values
df = df.fillna(method='ffill')  # Forward fill missing values

# Create features and target variables
df['Target'] = df['Close'].shift(-1)
df.dropna(inplace=True)
X = df[['Close']]
y = df['Target']
```

# Splitting the Data into Training and Testing Sets
We will split the dataset into training and testing sets for training and evaluating our machine learning model.
We split the dataset into features (X) and target variable (y) using data.drop() and data['Close'] respectively

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Scale the features
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train the model (Linear Regression)
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Make predictions
y_pred = model.predict(X_test_scaled)

# Convert y_test and y_pred to pandas Series
y_test_series = pd.Series(y_test, index=X_test.index)
y_pred_series = pd.Series(y_pred, index=X_test.index)

# Plot the predicted values and actual values
plt.figure(figsize=(12, 6))
plt.plot(y_test_series.index, y_test_series, label='Actual')
plt.plot(y_pred_series.index, y_pred_series, label='Predicted')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Actual vs Predicted Stock Prices')
plt.legend()
plt.show()
```

```python
# Calculate and print the root mean squared error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
```
